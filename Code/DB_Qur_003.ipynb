```python
def get_analysis(df, aya_column):
    letters = []

    for i in range(6236):
        aya_info = df.iloc[i]
        aya_text = aya_info[aya_column]
        aya_text = aya_text.replace(u'\xa0', u' ')  #.encode('utf-8')
        analysis = dict(Counter(aya_text))
        analysis['sura'] = aya_info['sura_number']
        analysis['aya'] = aya_info['aya_number_in_sura']
        letters.append(analysis)
    ds = pd.DataFrame(letters)
    ds = ds.reindex(sorted(ds.columns), axis=1).fillna(0)
    try:
        del (ds['\u200e'])
    except:
        pass
    ds['sum_letters'] = ds.iloc[:, 3:].sum(axis=1)
    ds['num_words'] = ds[' '] + 1

    # ds['alef'] = ds['ا'] + ds['أ'] + ds['إ'] + ds['آ'] + ds['ى']
    def safe_add(dataframe, cols, sum_name):
        dataframe[sum_name] = 0
        for a in cols:
            if a in dataframe.columns:
                dataframe[sum_name] = dataframe[sum_name] + dataframe[a]
    safe_add(ds, 'اأٱإٰآى', 'alef')
    safe_add(ds, 'ءٔئؤآ', 'hamza')
    safe_add(ds, 'ةت', 'all_ta')
    safe_add(ds, 'يۦ', 'all_ya')
    safe_add(ds, 'ۥو', 'all_waw')
    return ds
```

```python
ds_plain = get_analysis(df, 'aya_text_plain')
ds_plain.head()
```

```python
ds_plain_per_sura = ds_plain.groupby(['sura']).sum().drop(['aya', ' '], axis=1)
ds_plain_per_sura
```

```python
ds_uthmani = get_analysis(df, 'aya_text_uthmani')
```

```python
ds_uthmani_per_sura = ds_uthmani.groupby(['sura']).sum().drop(['aya', ' '], axis=1)
ds_uthmani_per_sura
```

```python
ds_tashkeel = get_analysis(df, 'aya_text_tashkeel')
ds_tashkeel_per_sura = ds_tashkeel.groupby(['sura']).sum().drop(['aya', ' '], axis=1)
display(ds_tashkeel, ds_tashkeel_per_sura)
```

```python
names = [
    'ds_plain', 'ds_plain_per_sura', 'ds_tashkeel', 'ds_tashkeel_per_sura',
    'ds_uthmani', 'ds_uthmani_per_sura'
]

for i, dtfr in enumerate([
        ds_plain, ds_plain_per_sura, ds_tashkeel, ds_tashkeel_per_sura,
        ds_uthmani, ds_uthmani_per_sura
]):
    dtfr.to_csv(os.path.join(path, f'{names[i]}.csv'),encoding='utf-8-sig')
    print(f'The dataset {names[i]}.csv was exported successfully')
```
